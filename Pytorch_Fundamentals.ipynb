Finding mean,max,min,sum of matrices

```python
matrix=torch.arange(0,11,1,dtype=torch.float32)
print(matrix)
print(torch.min(matrix))
print(torch.max(matrix))
print(torch.sum(matrix))
print(torch.mean(matrix)) ##doesnot work for long datatypes so changed the datatype tp float32
```

```python
##finding the positional min and max
print(matrix.argmin()) #find the position
print(matrix.argmax())
```

Reshaping ,Stacking ,Squeezing  and unsqueezing tensors
Reshaping :reshapes the tensor to a defined shape
stacking: combine multiple tensors sidebyside/vertical stacks/sequence of stacks
Squeezing :removes all "1" dimensions to target tensor
unsqueeze: add 1 dimension to target is tensor
permute:return a view of the input with dimesions permuted



```python
#lets create a tensor

x=torch.arange(1.,10.)
print(x,x.shape)
```

```python
x_reshaped=x.reshape(9,1)
print(x_reshaped,x.shape)
```

```python
#change the view
z=x.view(1,9)
print(z,z.shape)
```

```python
#changing z will change x bcuz view of a tensor shares the same memory as the original input
z[:,0]=5
x,z
```

```python
##stack tensors
x_stacked=torch.stack([x,x],dim=1)
x_stacked
```

```python

##torch squeeze removes single dimensions from tensor
x=torch.rand(size=(2,2))
print(x)
x_squeeze=torch.squeeze(x)
x_squeeze
```

```python
##permute shuffles the indexs

x_original=torch.rand(size=(224,224,0))
x_original

x_permuted=x_original.permute(2,0,1)
x_permuted
```

Reproducibility

```python
seed=42 ## to reproduce same tensor
torch.manual_seed(seed)

random_tensor_a=torch.rand(size=(2,2))
print(random_tensor_a)
torch.manual_seed(seed)
random_tensor_b=torch.rand(size=(2,2))
print(random_tensor_b)

print(random_tensor_a==random_tensor_b)
```
